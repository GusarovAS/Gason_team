```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('housing_train.csv')
#test = pd.read_csv('housing_test.csv')
data.head()
```

```python
from sklearn.model_selection import train_test_split
numeric_cols = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 
               'median_income']
categorical_cols = ['ocean_proximity']
data.shape
```

```python
data_dropna = data.dropna()

X_numeric_dropna = data_dropna[numeric_cols]
X_cat_dropna = data_dropna[categorical_cols]
y = data_dropna['median_house_value']
data_dropna.shape
```

```python
from sklearn.feature_extraction import DictVectorizer as DV
encoder = DV(sparse = False)
encoded_data = encoder.fit_transform(X_cat_dropna.T.to_dict().values())
```

```python
(X_train_numeric_dropna, 
 X_test_numeric_dropna, 
 y_train, y_test) = train_test_split(X_numeric_dropna, y, 
                                     test_size=0.3)
(X_train_cat_dropna,
 X_test_cat_dropna) = train_test_split(encoded_data, 
                                   test_size=0.3, 
                                   random_state=0)
```

```python
data_train_dropna = np.hstack((X_train_numeric_dropna, X_train_cat_dropna))
data_test_dropna = np.hstack((X_test_numeric_dropna, X_test_cat_dropna))
print("train size: " + str(data_train_dropna.shape[0]) + " test size: " + str(data_test_dropna.shape[0]))
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(data_train_dropna, y_train)
```

```python
reg.score(data_test_dropna, y_test)
```

```python
reg.coef_
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train_numeric_dropna)
X_train_numeric_dropna_scaled = scaler.transform(X_train_numeric_dropna)
scaler.fit(X_test_numeric_dropna)
X_test_numeric_dropna_scaled = scaler.transform(X_test_numeric_dropna)
```

```python
data_train_dropna_scaled = np.hstack((X_train_numeric_dropna_scaled, X_train_cat_dropna))
data_test_dropna_scaled = np.hstack((X_test_numeric_dropna_scaled, X_test_cat_dropna))
```

```python
reg.fit(data_train_dropna_scaled, y_train)
```

```python
reg.score(data_test_dropna_scaled, y_test)
```

```python
reg.coef_
```
